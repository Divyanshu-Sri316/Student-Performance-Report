```python
# Splitting dataframe in Dependent and Independent variables
X = df[['Gender', 'AttendanceRate', 'StudyHoursPerWeek',
       'PreviousGrade', 'ExtracurricularActivities', 'ParentalSupport']].values
y = df['FinalGrade'].values
```

```python
# Training-Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=0.3, random_state=0)
```

```python
# Fitting Multiple Linear Regression to the Training set
from sklearn.linear_model import LinearRegression
multireg = LinearRegression()
```

```python
multireg.fit(X_train, y_train)
```

```python
# Predicting the values
y_pred = multireg.predict(X_train)
```

```python
y_train, y_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
print('Mean Absolute Error is ', mean_absolute_error(y_train, y_pred))
print('Mean Squared Error is ', mean_squared_error(y_train, y_pred))
print('R-squared Score is ', r2_score(y_train, y_pred))
```

#### Hyperparameter Tuning

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

# Define pipeline
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('dim_reduction', PCA(n_components=5, svd_solver='auto')),
    ('model', Ridge())
])

# Grid search for hyperparameter tuning
param_grid = {
    'dim_reduction__n_components': [2, 5, 6, 7, 10, 12],
    'model__alpha': [0.01, 0.1, 1, 10],
}

grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)

print("Best Parameters:", grid_search.best_params_)
print("Best Score:", -grid_search.best_score_)
```

```python
grid_search.predict(X_test)
```

```python
y_test
```

```python
print('Mean Absolute Error is ', mean_absolute_error(y_test, grid_search.predict(X_test)))
print('Mean Squared Error is ', mean_squared_error(y_test, grid_search.predict(X_test)))
print('R-squared Score is ', r2_score(y_test, grid_search.predict(X_test)))
```

### Hyperparameter Tuning for Ensemble

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, ExtraTreesRegressor, AdaBoostRegressor, BaggingRegressor
from sklearn.model_selection import GridSearchCV, train_test_split
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_squared_error

# Models and parameter grids
models = {
    'RandomForest': RandomForestRegressor(random_state=0),
    'GradientBoosting': GradientBoostingRegressor(random_state=0),
    'ExtraTrees': ExtraTreesRegressor(random_state=0),
    'Bagging': BaggingRegressor(random_state=0),
    'AdaBoost': AdaBoostRegressor(random_state=0),
    'XGBoost': XGBRegressor(random_state=0),
    'LightGBM': LGBMRegressor(random_state=0),
}

param_grids = {
    'RandomForest': {
        'n_estimators': [100, 200, 500],
        'max_depth': [10, 20, None],
        'min_samples_split': [2, 5],
        'min_samples_leaf': [1, 2],
        'bootstrap': [True, False],
    },
    'GradientBoosting': {
        'n_estimators': [100, 200],
        'learning_rate': [0.01, 0.1],
        'max_depth': [3, 5],
        'min_samples_split': [2, 5],
        'min_samples_leaf': [1, 2],
    },
    'ExtraTrees': {
        'n_estimators': [100, 200],
        'max_depth': [10, 20, None],
        'min_samples_split': [2, 5],
        'min_samples_leaf': [1, 2],
    },
    'Bagging': {
        'n_estimators': [10, 50],
        'max_samples': [0.5, 0.7],
        'max_features': [0.5, 0.7],
    },
    'AdaBoost': {
        'n_estimators': [50, 100],
        'learning_rate': [0.01, 0.1],
    },
    'XGBoost': {
        'n_estimators': [100, 200],
        'learning_rate': [0.01, 0.1],
        'max_depth': [3, 5],
        'subsample': [0.8, 1.0],
    },
    'LightGBM': {
        'n_estimators': [100, 200],
        'learning_rate': [0.01, 0.1],
        'num_leaves': [31, 50],
        'max_depth': [-1, 10],
        'feature_fraction': [0.8, 1.0],
    },
}

# Perform hyperparameter tuning
best_models = {}
for name, model in models.items():
    print(f"Starting hyperparameter tuning for {name}...")
    grid = param_grids[name]
    grid_search = GridSearchCV(model, grid, cv=3, scoring='neg_mean_squared_error', n_jobs=-1, verbose=2)
    grid_search.fit(X_train, y_train)
    print(f"Best parameters for {name}: {grid_search.best_params_}")
    print(f"Best score for {name}: {-grid_search.best_score_}")
    print(f"Best r2score for {name}: {r2_score(y_test, grid_search.predict(X_test))}")

# Evaluate on test set
for name, model in best_models.items():
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    print(f"Test MSE for {name}: {mse}")
```

### The most accurate regressor is ExtraTreesRegressor
